```python
import csv
import ast

data = []
for line in lines:
    try:
        entry = ast.literal_eval(line.strip())   
        t = entry['timestamp']
        timelape = f"{t[0]}-{t[1]}"
        data.append([timelape, entry['text']])
    except Exception as e:
        print(f"Error processing {line}: {e}")

with open("speech_to_text.csv", "w", newline="", encoding="utf-8") as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(["timelape", "n·ªôi dung"])
    writer.writerows(data)

print("CSV file created successfully.")
```
